# Building a Dashboard showcasing major metrics for Pakistan's Economy using the Panel Library

```python
import pandas as pd
import numpy as np
import panel as pn
pn.extension('tabulator')

import hvplot.pandas
```

```python
df=pd.read_csv('Pakistan Economy.csv')
df
```

```python
df.describe()
```

```python
df_test=df.drop('Government', axis=1)
df_test
```

```python
df_test.corr()
```

```python
#Make dataframe interactive

idf= df.interactive()
```

```python
# Creating list for checkbox options
checkbox_options = df['Government'].unique().tolist()
```

```python
# Define CheckboxGroup widget
checkbox_group = pn.widgets.CheckBoxGroup(name='Government', options=checkbox_options, sizing_mode='stretch_height')
```

```python
checkbox_group
```

```python
#Create a function that updates the GDP growth plot according to the selected checkbox value
@pn.depends(checkbox_group.param.value)
def update_gdp_pipeline(selected_values):
    gdp_pipeline = idf[idf['Government'].isin(selected_values)]
    gdp_plot = gdp_pipeline.hvplot(x='Year', y='GDP Growth Rate %', title='GDP Growth Rate', line_width=3, width=500, height=200 )
    return pn.Column(gdp_plot)

```

```python
# Create the app layout
app_layout = pn.Column(
    update_gdp_pipeline
)
```

```python
app_layout.servable()
```

```python
#Create a similar function for the export growth chart
@pn.depends(checkbox_group.param.value)
def update_export_pipeline(selected_values):
    export_pipeline = idf[idf['Government'].isin(selected_values)]
    export_plot=export_pipeline.hvplot(x='Year', y='Export Growth %', title='Exports Growth Rate', line_width=3, width=500, height=200, line_color='green')
    return pn.Column(export_plot)
```

```python
# Create the app layout
export_app_layout = pn.Column(
    update_export_pipeline
)
```

```python
export_app_layout.servable()
```

```python
#Create scatterplot for exchange rate vs exports
scatter_plot= idf.hvplot.scatter(x='USD/PKR', y='Exports Trillions PKR', c='Government',
                                 cmap='Category10', legend='bottom_right', title='Exports vs Exchange Rate', 
                                size=90, width=500, height=375)
scatter_plot
```

```python
#Debt/GDP bar graph dataframe prep 
idf_debt=idf[['Government', 'Debt/GDP']]
idf_debt= idf_debt.groupby('Government')['Debt/GDP'].mean()
idf_debt= idf_debt.rename('Average Debt/GDP').round(1).reset_index()
idf_debt= idf_debt.reindex([1,0,2])
idf_debt
```

```python
#Plot the debt bar chart
debt_plot= idf_debt.hvplot(kind='bar', x='Government', y='Average Debt/GDP', c='Government',
                                 cmap='Category10', title='Average Debt/GDP', 
                                size=90, width=600, height=350, ylim=(50,80))
debt_plot
```

```python
#Prep the data frame to create an average happiness rank indicator
happiness_rank_df=df[['Government', 'Happiness Rank']].dropna()
happiness_rank_df=happiness_rank_df.groupby('Government')['Happiness Rank'].mean()
happiness_rank_df=happiness_rank_df.rename('Average Happiness Rank').reset_index()
happiness_rank_df['Average Happiness Rank']= happiness_rank_df['Average Happiness Rank'].round(1)
happiness_rank_df
```

```python
#Create Radiobuttongroup widget options for indicator
options=df['Government'].unique().tolist()
radio_button_group= pn.widgets.RadioButtonGroup(name='Government', options=options, value=[options[0]], orientation='vertical')
radio_button_group
```

```python
#Function to update number indicator in response to radiobuttongroup
@pn.depends(radio_button_group.param.value)
def updated_number_indicator(selected_government):
    #Filter data based on selected option
    filtered_df=happiness_rank_df[happiness_rank_df['Government']== selected_government]
    happiness_rank=filtered_df['Average Happiness Rank'].values[0]
    number_indicator=pn.indicators.Number(name='Mean Happiness', value=happiness_rank)
    return number_indicator
```

```python
#Number indicator layout
indicator_layout= pn.Column(updated_number_indicator)
```

```python
indicator_layout.servable()
```

```python
#Prep dataframe to create an average competitiveness rank indicator
competitiveness_rank_df=df[['Government', 'Competitiveness Rank']]
competitiveness_rank_df=competitiveness_rank_df.dropna()
competitiveness_rank_df=competitiveness_rank_df.groupby('Government')['Competitiveness Rank'].mean()
competitiveness_rank_df=competitiveness_rank_df.rename('Average Competitiveness Rank').reset_index()
competitiveness_rank_df['Average Competitiveness Rank']= competitiveness_rank_df['Average Competitiveness Rank'].round(1)
competitiveness_rank_df
```

```python
#Create button and indicator for competitiveness rank
#Function to update number indicator
@pn.depends(radio_button_group.param.value)
def updated_number_indicator(selected_government):
    #Filter data based on selected option
    filtered_df=competitiveness_rank_df[competitiveness_rank_df['Government']== selected_government]
    competitiveness_rank=filtered_df['Average Competitiveness Rank'].values[0]
    number_indicator=pn.indicators.Number(name='Mean Competitiveness', value=competitiveness_rank)
    number_indicator.color= 'red'
    number_indicator.width = 300
    return number_indicator
```

```python
#Number indicator layout
competitive_indicator_layout= pn.Column(updated_number_indicator)
competitive_indicator_layout.servable()
```

```python
#Prep data frame for average corruption rank indicator
corruption_rank_df=df[['Government', 'Corruption Rank']]
corruption_rank_df=corruption_rank_df.dropna()
corruption_rank_df=corruption_rank_df.groupby('Government')['Corruption Rank'].mean()
corruption_rank_df=corruption_rank_df.rename('Average Corruption Rank').reset_index()
corruption_rank_df['Average Corruption Rank']= corruption_rank_df['Average Corruption Rank'].round(1)
corruption_rank_df
```

```python
#Create button and indicator for corruption rank
#Function to update number indicator
@pn.depends(radio_button_group.param.value)
def updated_number_indicator(selected_government):
    #Filter data based on selected option
    filtered_df=corruption_rank_df[corruption_rank_df['Government']== selected_government]
    corruption_rank=filtered_df['Average Corruption Rank'].values[0]
    number_indicator=pn.indicators.Number(name='Mean Corruption', value=corruption_rank)
    number_indicator.color= 'red'
    return number_indicator
```

```python
#Number indicator layout
corruption_indicator_layout= pn.Column(updated_number_indicator)
corruption_indicator_layout.servable()
```

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='Pakistan Economic Performance Dashboard (2008-2022)', 
    sidebar=[pn.pane.JPG("vecteezy_pakistan-realistic-waving-flag-vector-illustration-national_7098518.jpg", sizing_mode='scale_both'),
             pn.pane.Markdown("# Pakistan Economic Performance by Government"), 
             pn.pane.Markdown("#### This dashboard makes it easy to sift through some key economic indicators like GDP growth, Export growth, and a few other indexes that reflect Pakistan's economic performance in the last few years."), 
             pn.pane.Markdown("### Control Index Indicators"),
             radio_button_group,
             pn.pane.Markdown("### Control Graphs"),
             checkbox_group,
             ],
    main=[pn.Row(
            indicator_layout,
            competitive_indicator_layout,
            corruption_indicator_layout,
            radio_button_group,
        ),
        pn.Row(
    app_layout, export_app_layout, checkbox_group
),
            pn.Row(scatter_plot.panel(width=50), debt_plot
        ),
        
    ],
    accent_base_color="#00A170",
    header_background="#00A170",
)
#template.show() opens the dashboard in a new window
template.show()
```
